```python
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

### First we need data (dimension):
- X data
- y data

```python
x = list(range(-10, 11, 1))
```

```python
y = []
for i in x:
    y.append(i**2)
    
#1. Empty List
#2. For Loop
#3. Append
```

```python
#Alternate approach: List comprehension
y = [i**2 for i in x]
```

```python
# vals = []
# for i in range(-100, 110, 10):
#     vals.append(i/100)
    
```

```python
len(x) == len(y)
```

```python
### "Stateful" / MATLAB-like syntax
plt.figure(figsize=(12, 8))
plt.plot(x, y, color='green')
plt.title('Polynomial $y = x^2$')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y'])
plt.grid()
plt.savefig('image.png')
```

```python
### "Pythonic" / Object-Oriented Method
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, y)
ax.set_title('Polynomial')
```

```python
def plot_data(x, y):
    plt.figure(figsize=(12, 8))
    plt.plot(x, y, color='green')
    plt.title('Polynomial $y = x^2$')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend(['y'])
    plt.grid()
```

```python
plot_data(x, y)
```
